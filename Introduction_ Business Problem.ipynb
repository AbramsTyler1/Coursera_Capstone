# Introduction/Business Problem

One of my friends who is a dentist wants to open her own dental clinic. Currently, her residence is in Germantown, Maryland, USA. And, she would like to open her dental clinic near to her residence. Also, she prefers to open her clinic in an area where not a lot of other dentists are practicing. By having lesser competition, she thinks her new business would have an edge, and higher success rate. Additionally, she would like to open her clinic near a public destination that is regularly visited by her prospective customers. For example, near a local grocery shop, movie theater, coffee shop, etc. Close proximity to one of these destinations would give her new dental clinic more visibility.
In order to solve the problem, I will need to find a suitable location for opening a new dental clinic near Germantown, Maryland. Any other dentist who is interested to open a clinic near Germantown (Maryland) may also benefit from the solution.
